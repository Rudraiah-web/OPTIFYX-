```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv(r"C:\Users\rudra\OneDrive\Desktop\Optifyx Technology\Iris.csv")
```

```python
print(data.head())
print(data.shape)
print(data.isnull().sum())
print(data.describe())
print(data.info())
print("Data Size:", data.size)
```

```python
data.drop(columns='Id', inplace=True)
```

```python
data_split = data.Species.str.split('-', expand=True)
data['Species'] = data_split[1] if len(data_split.columns) > 1 else data['Species']
```

```python
print("Data types:\n", data.dtypes)
print(data.isnull().sum())
```

```python
print("Species Counts:\n", data['Species'].value_counts())
```

```python
sns.FacetGrid(data, hue="Species", height=7).map(plt.scatter, "SepalLengthCm", "PetalLengthCm").add_legend()
plt.show()
```

```python
X = data[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]
y = data["Species"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=200)
model.fit(X_train, y_train)
```

```python
score = round(model.score(X_test, y_test), 4)

print("Model Accuracy:", score)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
sample_predictions = model.predict([[4.6, 3.2, 1.4, 0.2], [5.8, 2.7, 4.1, 1], [7.7, 2.8, 6.7, 2], [4.5, 8, 4.5, 2], [1.9, 4.2, 2.7, 1], [4.1, 6, 2, 3], [1.0, 2.5, 6, 0.4]])
print("Sample Predictions:", sample_predictions)
```
